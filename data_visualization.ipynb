```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("books_data.csv")

# Display first few rows
print("âœ… Dataset loaded successfully!")
data.head()
```

```python
# Check dataset info
print("\nðŸ“‹ Dataset Info:")
print(data.info())

# Summary statistics
print("\nðŸ“Š Summary Statistics:")
print(data.describe())
```

```python
# Convert 'Rating' to numeric (replace words like 'Three' -> 3)
rating_map = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
data['Rating_Num'] = data['Rating'].map(rating_map)

# Drop rows with missing values
data.dropna(inplace=True)

print("\nâœ… Cleaned Data Sample:")
data.head()
```

```python
# Distribution of Book Prices
plt.figure(figsize=(8,5))
sns.histplot(data['Price (Â£)'], bins=20, kde=True)
plt.title("Distribution of Book Prices")
plt.xlabel("Price (Â£)")
plt.ylabel("Count")
plt.show()
```

```python
# Count of Books by Rating
plt.figure(figsize=(7,5))
sns.countplot(x='Rating_Num', data=data, palette='viridis')
plt.title("Count of Books by Rating")
plt.xlabel("Rating (1â€“5 Stars)")
plt.ylabel("Number of Books")
plt.show()
```

```python
# Average Price per Rating
plt.figure(figsize=(8,5))
sns.barplot(x='Rating_Num', y='Price (Â£)', data=data, palette='coolwarm')
plt.title("Average Book Price by Rating")
plt.xlabel("Rating (1â€“5 Stars)")
plt.ylabel("Average Price (Â£)")
plt.show()
```

```python
# Top 10 Most Expensive Books
top10 = data.sort_values(by='Price (Â£)', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='Price (Â£)', y='Title', data=top10, palette='magma')
plt.title("Top 10 Most Expensive Books")
plt.xlabel("Price (Â£)")
plt.ylabel("Book Title")
plt.show()
```
