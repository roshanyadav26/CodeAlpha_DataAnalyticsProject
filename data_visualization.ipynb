```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("books_data.csv")

# Display first few rows
print("✅ Dataset loaded successfully!")
data.head()
```

```python
# Check dataset info
print("\n📋 Dataset Info:")
print(data.info())

# Summary statistics
print("\n📊 Summary Statistics:")
print(data.describe())
```

```python
# Convert 'Rating' to numeric (replace words like 'Three' -> 3)
rating_map = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
data['Rating_Num'] = data['Rating'].map(rating_map)

# Drop rows with missing values
data.dropna(inplace=True)

print("\n✅ Cleaned Data Sample:")
data.head()
```

```python
# Distribution of Book Prices
plt.figure(figsize=(8,5))
sns.histplot(data['Price (£)'], bins=20, kde=True)
plt.title("Distribution of Book Prices")
plt.xlabel("Price (£)")
plt.ylabel("Count")
plt.show()
```

```python
# Count of Books by Rating
plt.figure(figsize=(7,5))
sns.countplot(x='Rating_Num', data=data, palette='viridis')
plt.title("Count of Books by Rating")
plt.xlabel("Rating (1–5 Stars)")
plt.ylabel("Number of Books")
plt.show()
```

```python
# Average Price per Rating
plt.figure(figsize=(8,5))
sns.barplot(x='Rating_Num', y='Price (£)', data=data, palette='coolwarm')
plt.title("Average Book Price by Rating")
plt.xlabel("Rating (1–5 Stars)")
plt.ylabel("Average Price (£)")
plt.show()
```

```python
# Top 10 Most Expensive Books
top10 = data.sort_values(by='Price (£)', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='Price (£)', y='Title', data=top10, palette='magma')
plt.title("Top 10 Most Expensive Books")
plt.xlabel("Price (£)")
plt.ylabel("Book Title")
plt.show()
```

```python
# Insights Summary
print("📈 Insights:")
print("1️⃣ Most books are priced under £40.")
print("2️⃣ Books rated 4–5 stars tend to have slightly higher average prices.")
print("3️⃣ Price distribution is slightly right-skewed, indicating a few premium-priced books.")
print("4️⃣ Top 10 most expensive books can be highlighted for further analysis or marketing insights.")
```
